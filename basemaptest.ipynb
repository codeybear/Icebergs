```python
plt.figure(figsize=(8,8))
m = Basemap(projection='aeqd', resolution='l', lat_0=-90, lon_0=-138, width=10000000, height=10000000)
            #llcrnrx=0, llcrnry=0, urcrnrx=10000, urcrnry=10000)
            #llcrnrlon=0.,llcrnrlat=-60.,urcrnrlon=-180.,urcrnrlat=-60.)
m.drawmeridians(np.arange(0,360,30))
m.drawparallels(np.arange(-90,90,30))

lons = df['Longitude'].values
lats = df['Latitude'].values
iceberg = df['Iceberg'].values

x, y = m(lons, lats)

m.scatter(x, y, 3, marker='D', color='m')
m.bluemarble(scale=0.5)
```
